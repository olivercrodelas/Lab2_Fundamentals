```python
# [CELL 1] CONFIGURATION
ENGINEER = "Rodelas"
STUDENT_ID = "TUPM-25-0512"

# Derivation: Extract the last digit to act as your Logic Seed
SEED_NUM = int(STUDENT_ID[-1])

print(f"System Initialized for: {ENGINEER}")
print(f"Algorithmic Seed Value: {SEED_NUM}")
```

```python
# [CELL 2] ARCHITECTURE TEST
# Loic: Mathematically, 0.1 + 0.2 = 0.3. Does the computer agree?
actual = 0.1 + 0.2

print(f"Python stores: {actual}")

if actual ==0.3:
    print("Result: Perfect Match")
else:
    print("Result: PRECISION ERROR (Binary Artifact)")
    
```

```python
# [CELL 3] ASCII INSPECTION
first_letter = ENGINEER[0]
byte_val = ord(first_letter) # Convert Char to Integer Memory Code

print(f"Character: {first_letter}")
print(f"ASCII Value: {byte_val}")
print(f"Next in Sequence: {chr(byte_val + 1)}")
```

```python
# [CELL 4] SANITIZATION
raw_in = input("Enter Name (add messy spaces)")
# Pipeline: Strip whitespace -> Convert Uppercase
clean_in = raw_in.strip().upper()
print(f"Original: '{raw_in}' | Cleaned: '{clean_in} ")
```

```python
# [CELL 5] DATA EXISTENCE
user_input = input("Press Enter (leave empty) or type data: ")

# Pythonic Way: An empty string is treated as False.
if user_input:
    print(f"Data Detected: {user_input}")
else:
    print("ALERT: No data stream detected.")
```

```python
# [CELL 6] SLICING & PARSING
# Using the STUDENT_ID defined in Task 4 (e.g., "TUPM-24-1234")
campus = "TUPM-25-0512"[0:4]
unique_id = "TUPM-25-0512"[8:]

# Tokenization: Split string into a List based on delimiter "-"
tokens = STUDENT_ID.split("-")

print(f"Campus Code: {campus}")
print(f"Unique Sequence: {unique_id}")
print(f"Token List: {tokens}")
```

```python
# [CELL 7] MODULO OPERATOR
# Logic: Check if Seed Number is Even or Odd using %
if SEED_NUM % 2 == 0:
    priority = "LOW (Even Seed)"
else:
    priority = "HIGH (Odd Seed)"

print(f"System Priority: {priority}")
```

```python
# [CELL 8] DYNAMIC LIMITS
# The limit changes for every student
limit = (SEED_NUM * 10) + 20
print(f"--- SAFETY LIMIT: {limit} ---")

val = int(input("Enter Sensor Reading: "))

if val > limit:
    print("ALERT: Over Limit")
elif val < 5:
    print("ALERT: Low Signal")
else:
    print("STATUS: Nominal")
```

```python
# [CELL 9] ROBUST INPUT (Exception Handling)
try:
    # Attempt to convert input to integer
    voltage = int(input("Enter Voltage (Numeric): "))
    print(f"Voltage Captured: {voltage}V")
    
except ValueError:
    # This block runs ONLY if the user types text (e.g., "five") 
    # or leaves it blank, which normally causes a crash.
    print("SYSTEM ERROR: Input must be numeric. Resetting to 0.")
    voltage = 0

# The program remains stable and continues to this line
print(f"System continues... Current Voltage: {voltage}")
```

```python
# [CELL 10] LOGIC GATES
has_power = False

# Python sees False and STOPS checking. It never runs the second part.
if has_power and (10 / 0 == 0):
    print("System On")
else:
    print("System Off (Short-Circuit Triggered)")
```

```python
# [CELL 11] FOR LOOP
loops = SEED_NUM + 3
print(f"--- RUNNING {loops} CYCLES ---")  

for i in range(1, loops + 1): 
    print(f"Cycle {i}/{loops} ... Complete") 
```

```python
# [CELL 12] NESTED LOOPS
# First digit of the unique ID token determines Rows
rows = int(tokens[2][0]) 
# Last digit of the unique ID token determines Columns
cols = int(tokens[2][-1]) 

print(f"Generating {rows}x{cols} Matrix:")

for r in range(rows):
    for c in range(cols):
        print("#", end=" ") 
    print()
```

```python
# [CELL 13] INPUT VALIDATION
while True:
    pwd = input("Type 'ADMIN' to exit: ")
    
    if pwd == "ADMIN":
        print("Unlocked.")
        break # This command terminates the infinite loop
    else:
        print("Access Denied.")
```

```python
# [CELL 14] FORENSIC HASH CALCULATION
# A = ASCII Value of first letter (Calculated in Task 5)
A = byte_val 

# B = Seed Number (Calculated in Task 4)
B = SEED_NUM 

# C = Total Loop Cycles (Calculated in Task 14)
C = loops 

# Formula: (ASCII * Seed) - Loops
verification_hash = (A * B) - C

print(f"User: {ENGINEER}")
print(f"Forensic Signature: {verification_hash}")
```
